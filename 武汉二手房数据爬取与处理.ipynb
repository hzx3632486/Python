```python
import requests
from bs4 import BeautifulSoup
import pandas
```

```python
houseary = []
priceary = []
```

```python
for i in range(1):
    res = requests.get('https://wh.lianjia.com/ershoufang/pg'+str(i))
    soup = BeautifulSoup(res.text,'html.parser')
    for house in soup.select('.sellListContent li'):
        houseary.append({'亮点':house.select('a')[1].text,'小区':house.select('a')[2].text,'区域':house.select('a')[3].text,
                         '链接':house.select('a')[0]['href']})
    for price in soup.select('.priceInfo'):
        priceary.append({'zongjia':price.select('span')[0].text,'danji':price.select('span')[1]})
```

```python
houseary
```

```python
housedf = pandas.DataFrame(houseary)
pricedf = pandas.DataFrame(priceary)
```

```python
pricedf
```

```python
frames = [housedf, pricedf]
```

```python
result = pandas.concat(frames)
```
